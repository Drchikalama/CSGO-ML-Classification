## Predicting which team will win a CS:GO round

### Data Aquiosition

```python
import pandas as pd
```

```python
df = pd.read_csv("csgo_round_snapshots.csv")
```

```python
df.head()
```

### Data Cleaning and EDA

```python
df.columns # Too many columns to consider
```

```python
important_cols = ['time_left','ct_score','t_score','map','bomb_planted','ct_health','t_health',
                  'ct_armor','t_armor','ct_money','ct_players_alive','t_players_alive','round_winner',]

df = df[important_cols].copy()

df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[df['t_players_alive']==6] # Each team is supposed to have only 5 members.
```

```python
df.drop(index = 58271, inplace = True) 
```

```python
df.columns
```

```python
df.drop(columns = ['map','round_winner','bomb_planted'], axis = 1).corr() > 0.9

# We can see that the only highly corrolated features are the health and players alove for each team.
# Therefore, using both of them is not good for the model.
```

```python
df.drop(columns = ['ct_players_alive','t_players_alive'], axis = 1, inplace = True)
```

```python
df.round_winner.replace(to_replace='CT', value=1, inplace=True) # Replacing CT win with 1
df.round_winner.replace(to_replace='T', value=0, inplace=True) # Replacing T win with 0

df.bomb_planted.replace({True: 1, False: 0}, inplace=True) # Replacing boolean values with integer
```

```python
df.head()
```

### Preprocessing

##### Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc = OrdinalEncoder()
# Repalcing string of map name with a number for each map
df['map'] = enc.fit_transform(df['map'].values.reshape(-1, 1))
```

```python
df.head()
```

##### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

columns = df.columns
target = df['round_winner'].copy()

df = pd.DataFrame(scaler.fit_transform(df) ,columns=columns)
df['round_winner']=target

df.dropna(inplace=True)
```

```python
df
```

##### Splitting the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns = 'round_winner')
y = df['round_winner']

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1)
```

### Applying machine learning algorithems for classification

##### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_clf = LogisticRegression(random_state=0)

lr_model = lr_clf.fit(X_train, y_train)

lr_score = lr_clf.score(X_test, y_test)
print (lr_score)
```

##### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(n_estimators=100, max_depth=None)

rf_model = rf_clf.fit(X_train, y_train)

rf_score = rf_clf.score(X_test, y_test)
print (rf_score)
```

##### Naive Bayes

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import Normalizer
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
NB_clf = GaussianNB()

NB_model = NB_clf.fit(X_train, y_train)

NB_score = NB_clf.score(X_test, y_test)
print(NB_score)
```

##### SVC

```python
from sklearn.linear_model import SGDClassifier
```

```python
SVC_clf = SGDClassifier(random_state=0)

SVC_model = SVC_clf.fit(X_train, y_train)

SVC_score = SVC_clf.score(X_test, y_test)
print(SVC_score)
```
